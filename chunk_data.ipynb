```python
import xarray as xr

ds = xr.open_dataset("./data/2m_temperature-2020.nc")
ds
```

```python
import numpy as np

# every 24th hour of ds.time
for t in ds.time[::24]:
    date = str(t.values)[:10]
    ds_single = ds.sel(time=t)
    ds_single.to_netcdf(f"./data/2m_temperature-{date}.nc")
```
